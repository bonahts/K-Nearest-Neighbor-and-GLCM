```python
import sys
#pandas untuk manipulasi data 
import pandas as pd 
import cv2 as cv
import imutils.paths as path 
import numpy as np
from tqdm import tqdm
from skimage.feature import greycomatrix, greycoprops
#library matplotlib untuk grafik 
import matplotlib.pyplot as plt
import seaborn as sns
import os.path
from sklearn.model_selection import train_test_split
#sklearn untuk knn classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python

def calc_glcm_all_agls(img, label, props, dists=[1], agls=[0, np.pi/4, np.pi/2, 3*np.pi/4], lvl=256, sym=True, norm=True):
    
    glcm = greycomatrix(img, 
                        distances=dists, 
                        angles=agls, 
                        levels=lvl,
                        symmetric=sym, 
                        normed=norm)
    feature = []
    glcm_props = [propery for name in props for propery in greycoprops(glcm, name)[0]]
    for item in glcm_props:
            feature.append(item)
    feature.append(label) 
    
    return feature
```

```python
PATH = 'D:\SKRIPSI\data set daun\daun mangga keweni_r'  
imagePaths = sorted(list(path.list_images(PATH))) 

#untuk menyimpan citra dan mengurutkan citra
```

```python
data = []
#properties = ['dissimilarity', 'correlation', 'homogeneity', 'contrast', 'ASM', 'energy'] 
properties = ['contrast','dissimilarity','homogeneity','ASM','energy']

label = 0 #keweni
glcm_all_agls = []
for i in tqdm(imagePaths,desc="load"):
    img=cv.imread (i)
    imgresize = cv.resize (img, (261,141))
    hasilgrayscale = cv.cvtColor(imgresize, cv.COLOR_BGR2GRAY)


    glcm_all_agls.append(calc_glcm_all_agls(hasilgrayscale, label, props=properties))

    # cv.imshow(img)
```

```python
PATH_lalijowo = 'D:\SKRIPSI\data set daun\daun mangga lalijiwo_r'
imagePaths_lalijowo = sorted(list(path.list_images(PATH_lalijowo)))
```

```python
label_lalijowo = 1 #lalijowo
for i in tqdm(imagePaths_lalijowo,desc="load"):
    img=cv.imread (i)
    imgresize = cv.resize (img, (261,141))
    hasilgrayscale = cv.cvtColor(imgresize, cv.COLOR_BGR2GRAY)

    glcm_all_agls.append(calc_glcm_all_agls(hasilgrayscale, label_lalijowo, props=properties))
```

```python
PATH_madu = 'D:\SKRIPSI\data set daun\daun mangga madu_r'
imagePaths_madu = sorted(list(path.list_images(PATH_madu)))
```

```python
label_madu = 2 #madu
for i in tqdm(imagePaths_madu,desc="load"):
    img=cv.imread (i)
    imgresize = cv.resize (img, (261,141))
    hasilgrayscale = cv.cvtColor(imgresize, cv.COLOR_BGR2GRAY)
    #(thresh, hasil_biner) = cv.threshold(hasilgrayscale, 150, 256, cv.THRESH_BINARY | cv.THRESH_OTSU)

    glcm_all_agls.append(calc_glcm_all_agls(hasilgrayscale, label_madu, props=properties))
```
