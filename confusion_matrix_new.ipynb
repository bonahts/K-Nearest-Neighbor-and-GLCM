```python
## COBA KNN
# ACAK DATASET
from sklearn.utils import shuffle

df = shuffle(glcm)
df
```

```python
# #create a dataframe with all training data except the target column
x = df.drop(columns=['label'])

# #check that the target variable has been removed
x.head()
```

```python
#separate target values
y = df['label'].values

#view target values
y
```

```python
type(x)
```

```python
x = x.to_numpy()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.metrics import confusion_matrix
from statistics import mean 

labels = [0,1,2]
acc=[]
presisi=[]
recal=[]

tprs = []

fold = 1

skfold = StratifiedKFold(n_splits=5, shuffle = True)

for i, (train, test) in enumerate(skfold.split(x, y), start=1):
    print('fold : ', i)
    x_train, x_test = x[train], x[test]
    y_train, y_test = y[train], y[test]
    # Create KNN classifier
    knn = KNeighborsClassifier(n_neighbors = 7)

    # Fit the classifier to the data
    knn.fit(x_train,y_train)
    
    result = knn.predict(x_test)
    
    print(y_test)
    print(result)

    cf_klasifikasi = pd.DataFrame(data=confusion_matrix(y_test, result, labels=labels), columns=labels, index=labels)
    print(cf_klasifikasi)
    
    tps_klasifikasi = {}
    fps_klasifikasi  = {}
    fns_klasifikasi  = {}
    tns_klasifikasi  = {}
    for label in labels:
        tps_klasifikasi[label] = cf_klasifikasi.loc[label, label]
        fps_klasifikasi[label] = cf_klasifikasi[label].sum() - tps_klasifikasi[label]
        fns_klasifikasi[label] = cf_klasifikasi.loc[label].sum() - tps_klasifikasi[label]

    for label in set(y_test):
        tns_klasifikasi[label] = len(y_test) - (tps_klasifikasi[label] + fps_klasifikasi[label] + fns_klasifikasi[label])

    accuracy_global_new_klasifikasi=sum(tps_klasifikasi.values())/len(y_test)
    acc.append(accuracy_global_new_klasifikasi)


    tpfp_klasifikasi = [ai + bi for ai, bi in zip(list(tps_klasifikasi.values()), list(fps_klasifikasi.values()))]
    precision=[ai / bi if bi>0 else 0 for ai , bi in zip(list(tps_klasifikasi.values()), tpfp_klasifikasi)]
    precisionklasifikasi=sum(precision)/3
    presisi.append(precisionklasifikasi)

    tpfn_klasifikasi = [ai + bi for ai, bi in zip(list(tps_klasifikasi.values()), list(fns_klasifikasi.values()))]
    recall=[ai / bi if bi>0 else 0 for ai, bi in zip(list(tps_klasifikasi.values()), tpfn_klasifikasi)]
    recallklasifikasi=sum(recall)/3
    recal.append(recallklasifikasi)

    print('akurasi : ',accuracy_global_new_klasifikasi)
    print('presisi : ',precisionklasifikasi)
    print('recall : ',recallklasifikasi,'\n')
print('###############################################\n')    
print('rata2 akurasi : ', mean(acc))
print('rata2 presisi : ', mean(presisi))
print('rata2 recal : ', mean(recal))
```
